La función **heatmap** permite crear mapas de calor o gráficos de correlación en Python con seaborn. Puedes pasar un conjunto de datos 2D o un data frame de pandas. 

En caso de utilizar un data frame de pandas los nombres de las filas y las columnas serán usados como etiquetas para los ejes.

```python
import numpy as np
import seaborn as sns

# Simulación de datos
np.random.seed(1)
data = np.random.rand(10, 10)

sns.heatmap(data)
```

```python
#Celdas cuadradas
import numpy as np
import seaborn as sns

np.random.seed(1)
data = np.random.rand(10, 10)

sns.heatmap(data, square = True)
```

```python
#Borde
import numpy as np
import seaborn as sns

np.random.seed(1)
data = np.random.rand(10, 10)

sns.heatmap(data,
            linewidths = 0.75, 
            linecolor = "white") # Color por defecto
```

```python
#Anotaciones de texto
import numpy as np
import seaborn as sns

np.random.seed(1)
data = np.random.rand(10, 10)

sns.heatmap(data,
            annot = True,
            annot_kws = {'size': 6}) # Argumentos adicionales 
```

```python
#Personalizacion de color
import numpy as np
import seaborn as sns

np.random.seed(1)
data = np.random.rand(10, 10)

sns.heatmap(data,
            cmap = "Blues")
```

```python
#Challenge: 

# A partir de este conjunto de datos, propuesto. Realizar las consignas solicitadas:

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Estudiante': ['Estudiante1', 'Estudiante2', 'Estudiante3', 'Estudiante4', 'Estudiante5'],
    'Matematicas': [85, 78, 92, 88, 95],
    'Ciencias': [75, 89, 82, 93, 87],
    'Historia': [68, 72, 64, 78, 82]
}

df = pd.DataFrame(data)
```

```python
#1. Consigna: Crear un histograma de las puntuaciones en Matematicas.
sns.histplot(df['Matematicas'], bins=5, kde=True)
plt.title("Histograma de Puntuaciones en Matematicas")
plt.xlabel("Puntuación en Matematicas")
plt.ylabel("Frecuencia")
plt.show()
```

```python
#2. Consigna: Crear un gráfico de barras de las puntuaciones promedio en Ciencias por estudiante.
sns.barplot(x='Estudiante', y='Ciencias', data=df)
plt.title("Puntuaciones Promedio en Ciencias por Estudiante")
plt.xlabel("Estudiante")
plt.ylabel("Puntuación Promedio en Ciencias")
plt.xticks(rotation=45)
plt.show()
```

```python
#3. Consigna: Crear un gráfico de dispersión de las puntuaciones en Historia vs. Ciencias.
sns.scatterplot(x='Ciencias', y='Historia', data=df)
plt.title("Puntuaciones en Historia vs. Ciencias")
plt.xlabel("Puntuación en Ciencias")
plt.ylabel("Puntuación en Historia")
plt.show()
```

```python
# 4. Consigna: Calcular y mostrar la matriz de correlación entre todas las materias.
correlation_matrix = df[['Matematicas', 'Ciencias', 'Historia']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Matriz de Correlación entre Materias")
plt.show()
```

```python
# 5. Consigna: Crear un boxplot de las puntuaciones en Matematicas.
sns.boxplot(x='Matematicas', data=df)
plt.title("Boxplot de Puntuaciones en Matematicas")
plt.xlabel("Puntuación en Matematicas")
plt.show()
```
