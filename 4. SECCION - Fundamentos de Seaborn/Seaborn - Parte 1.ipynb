**Seaborn** es una biblioteca de visualización de datos en Python que se basa en Matplotlib y proporciona una interfaz de alto nivel para crear visualizaciones estadísticas atractivas y efectivas. Fue desarrollada para simplificar y mejorar la creación de gráficos en Python al ofrecer una sintaxis más simple y amigable, así como una mayor estilización por defecto en comparación con Matplotlib. 

Para utilizar Seaborn! debemos instalar la librería utilizando: 

* pip install seaborn 
* conda install seaborn

Veamos cómo hacerlo! :)

```python
#Empecemos viendo el ejemplo más sencillo, un histograma!:

import numpy as np
import seaborn as sns
 
sns.set(style="white")
 
# Generate a random univariate dataset
rs = np.random.RandomState(10)
d = rs.normal(size=100)
 
# Plot a simple histogram and kde
sns.histplot(d, kde=True, color="m")
```

```python
#Veamos un ejemplo un poco mas! complejo
import seaborn as sns
 
sns.set(style="dark")
fmri = sns.load_dataset("fmri")
 
# Plot the responses for different\
# events and regions
sns.lineplot(x="timepoint",
             y="signal",
             hue="region",
             style="event",
             data=fmri)
```

```python
#Otro gráfico super habitual es ...
import seaborn as sns
 
sns.set(style="ticks")
 
# Loading the dataset
df = sns.load_dataset("anscombe")
 
# Show the results of a linear regression
sns.lmplot(x="x", y="y", data=df)
```

```python
#Excelente, ahora que tenemos un entendimiento básico de seaborn, veamos como
#empezar a customizar nuestras visuales!
```

```python
#Paletas de colores
from matplotlib import pyplot as plt
import seaborn as sns
current_palette = sns.color_palette()
sns.palplot(current_palette)
plt.show()
```

```python
#Paleta de grises
from matplotlib import pyplot as plt
import seaborn as sns
current_palette = sns.color_palette()
sns.palplot(sns.color_palette("Greys"))
plt.show()
```

```python
#Divergente
from matplotlib import pyplot as plt
import seaborn as sns
current_palette = sns.color_palette()
sns.palplot(sns.color_palette("terrain_r"))
plt.show()
```

```python
#Colores brillantes!
import pandas as pd
import seaborn as sns
  
sns.palplot(sns.color_palette("bright"))
```

```python
#Paleta oscura de colores
import pandas as pd
import seaborn as sns
  
sns.palplot(sns.color_palette("dark", 10))
```

```python
#Personalizando nuestra paleta de colores:
import pandas as pd
import seaborn as sns
  
color = ["green", "White", "Red", "Yellow", "Green", "Grey"]
sns.set_palette(color)
sns.palplot(sns.color_palette())
```
