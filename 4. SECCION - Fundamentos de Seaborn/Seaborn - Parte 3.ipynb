```python
#Grafico de linea
import seaborn as sns

# Gráfico    
ax = sns.lineplot(x = [1, 2, 3, 4, 5],
                  y = [1, 4, 2, 5, 4])

# Anotación de texto en el punto x = 3, y = 1.75           
ax.text(3, 1.75, "Texto personalizado")
```

```python
#Personalizacion de texto!
import seaborn as sns

# Gráfico    
ax = sns.lineplot(x = [1, 2, 3, 4, 5],
                  y = [1, 4, 2, 5, 4])

# Anotación de texto en el punto x = 3, y = 1.75           
ax.text(3, 1.75, "Texto personalizado", 
       fontsize = 12,          # Tamaño
       fontstyle = "oblique",  # Estilo
       color = "red",          # Color
       ha = "center", # Alineación horizontal
       va = "center") # Alineación vertical
```

```python
#Incorporamos una rotacion de texto
import seaborn as sns

# Gráfico    
ax = sns.lineplot(x = [1, 2, 3, 4, 5],
                  y = [1, 4, 2, 5, 4])

# Anotación de texto en el punto x = 2, y = 1           
ax.text(2, 1, "Texto personalizado", 
       rotation = -50) 
```

```python
#Etiquetar los puntos de los datos
import seaborn as sns

x = [1, 2, 3, 4, 5]
y = [1, 4, 2, 5, 4]
textos = ["A", "B", "C", "D", "E"]

# Gráfico    
ax = sns.scatterplot(x = x, y = y)

for i, txt in enumerate(textos):
    ax.text(x[i], y[i], txt)
```

```python
#Scatterplot!
import seaborn

seaborn.set(style='whitegrid')
fmri = seaborn.load_dataset("fmri")
 
seaborn.scatterplot(x="timepoint",
                    y="signal",
                    data=fmri)
```

```python
import seaborn
  
seaborn.set(style='whitegrid')
fmri = seaborn.load_dataset("fmri")
 
seaborn.scatterplot(x="timepoint",
                    y="signal",
                    hue="region",
                    style="event",
                    data=fmri)
```

```python
#Un derivado del gráfico de scatter, es ..

import seaborn as sns

df = sns.load_dataset("penguins")

sns.jointplot(data = df,
              x = "bill_length_mm",
              y = "bill_depth_mm")
```

```python
#Agrega recta de regresión y densidades
import seaborn as sns

df = sns.load_dataset("penguins")

sns.jointplot(data = df,
              x = "bill_length_mm",
              y = "bill_depth_mm",
              kind = "reg")  
```

```python
#Uso de marcadores
import seaborn as sns

df = sns.load_dataset("penguins")

sns.jointplot(data = df,
              x = "bill_length_mm",
              y = "bill_depth_mm", 
              marker = "x")
```

```python
#Barplot
import seaborn as sns
import matplotlib.pyplot as plt
 
df = sns.load_dataset('titanic')
 
sns.barplot(x = 'class', y = 'fare', data = df)
 
plt.show()
```

```python
#Parametro hue
import seaborn as sns

import matplotlib.pyplot as plt
 
df = sns.load_dataset('titanic')
 
sns.barplot(x = 'class', y = 'fare', hue = 'sex', data = df)
 
plt.show()
```

```python
#Rotacion del gráfico
import seaborn as sns

import matplotlib.pyplot as plt
 
df = sns.load_dataset('titanic')

sns.barplot(x = 'fare', y = 'class', hue = 'sex', data = df)
 
plt.show()
```

```python
#Ordenamiento
import seaborn as sns

import matplotlib.pyplot as plt
 
df = sns.load_dataset('titanic')

sns.barplot(x = 'class', y = 'fare', data = df,
            order = ["Third", "Second", "First"])
 
plt.show()
```

```python
#Uso de colores
import seaborn as sns

import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')
 
sns.barplot(x = 'class', y = 'fare', data = df, color = "salmon")
 
plt.show()
```

```python
#Uso de paleta de colores
import seaborn as sns

import matplotlib.pyplot as plt
 
df = sns.load_dataset('titanic')
 
sns.barplot(x = 'class', y = 'fare',
            hue = 'sex', data = df, palette='pastel')
 
plt.show()
```

```python
#Saturacion
import seaborn as sns

import matplotlib.pyplot as plt
 
df = sns.load_dataset('titanic')
 
sns.barplot(x = 'class', y = 'fare', hue = 'sex', data = df,saturation = 0.1)

plt.show()
```

```python
#Por ultimo! count plot
import seaborn as sns

data = ["A", "A", "B",
        "B", "B", "C"]

sns.countplot(x = data)
```

```python
#Count plot horizontal
import seaborn as sns

data = ["A", "A", "B",
        "B", "B", "C"]

sns.countplot(y = data)
```

```python
#Count plot por grupos y subgrupos
import seaborn as sns

data = ["A", "A", "B", "A",
        "B", "B", "C", "C"]
group = ["G1", "G1", "G2", "G2",
         "G1", "G2", "G1", "G2"]

sns.countplot(x = data, hue = group)
```

```python
#Definiendo un color de borde!
import seaborn as sns

data = ["A", "A", "B",
        "B", "B", "C"]

# Count plot
sns.countplot(x = data,
              edgecolor = "red")
```
