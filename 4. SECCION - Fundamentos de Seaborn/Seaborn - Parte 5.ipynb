Un **stripplot** es un tipo de gráfico de dispersión utilizado en estadísticas y visualización de datos para mostrar la distribución de puntos a lo largo de una sola variable categórica o de factor. 

Es especialmente útil cuando deseas visualizar la distribución de valores individuales en una variable categórica y entender cómo se agrupan o dispersan estos valores.

Veamos algunos ejemplos :)

```python
import numpy as np
import seaborn as sns
import random

# Simulación de datos
rng = np.random.RandomState(7)
variable = rng.normal(0.5, 2, size = 250)
random.seed(7)
grupo = random.choices(["G1", "G2", "G3"], k = 250)
grupo2 = random.choices(["A", "B"], k = 250)
df = {'variable': variable, 'grupo': grupo, 'grupo2': grupo2}
```

```python
import seaborn as sns

sns.stripplot(x = variable)
```

```python
#Vertical
import seaborn as sns

sns.stripplot(y = variable)
```

```python
#Color
import seaborn as sns

sns.stripplot(x = variable,
              color = "darkseagreen")
```

```python
#Borde
import seaborn as sns

sns.stripplot(x = variable,
              linewidth = 1,
              color = "darkseagreen")
```

```python
#Marker
import seaborn as sns

sns.stripplot(x = variable,
              marker = "D")
```

```python
#Tamaño y transparencia
import seaborn as sns

sns.stripplot(x = variable,
              size = 10,   # Tamaño
              alpha = 0.2) # Transparencia 
```

```python
#Strip plot por grupo
import seaborn as sns

sns.stripplot(x = variable,
              y = grupo)
```

```python
#Orientacion
import seaborn as sns

sns.stripplot(y = variable,
              x = grupo)
```

```python
#Ordenamiento
import seaborn as sns

sns.stripplot(y = variable,
              x = grupo,
              order = ["G3", "G2", "G1"])
```

El gráfico **Swarmplot** en Seaborn es un tipo de gráfico de dispersión que se utiliza para visualizar la distribución de valores en una variable categórica o de factor. Es similar al stripplot, pero con la diferencia clave de que los puntos en un swarmplot se ajustan automáticamente para evitar la superposición, lo que facilita la visualización de la densidad de datos y la identificación de patrones más claramente.

```python
import seaborn as sns

sns.swarmplot(x = variable)

# Equivalente a:
sns.swarmplot(x = "variable", data = df)
```

```python
#Swarm plot vertical
import seaborn as sns

sns.swarmplot(y = variable) 
```

```python
#Tamaño de los puntos
import seaborn as sns

sns.swarmplot(y = variable,
              size = 4) 
```

```python
#Personalizar el color de fondo de los puntos
import seaborn as sns

sns.swarmplot(y = variable,
              color = "green")
```

```python
#Grosor y color del borde de los puntos
import seaborn as sns

sns.swarmplot(y = variable,
              edgecolor = "red",
              linewidth = 0.5)
```

```python
#Por ultimo, veamos algunos ejemplos sobre: Swarm plot por grupo en seaborn
import seaborn as sns

# Swarm plot
sns.swarmplot(x = grupo, y = variable)

# Equivalent to:
sns.swarmplot(x = "grupo", y = "variable", data = df)
```

```python
#Orientacion
import seaborn as sns

sns.swarmplot(y = grupo, x = variable)
```

```python
#Orden personalizado
import seaborn as sns

sns.swarmplot(x = grupo, y = variable,
              order = ["G1", "G2", "G3"]) 
```
