Los gráficos de **FacetGrid** en Seaborn son una poderosa herramienta para visualizar datos que involucran múltiples variables categóricas. 

Estos gráficos permiten dividir un conjunto de datos en subconjuntos más pequeños según una o más variables categóricas y crear visualizaciones separadas para cada uno de estos subconjuntos. 

Esto facilita la comparación y la identificación de patrones o relaciones en los datos en función de las categorías de interés. 

```python
#Ejemplo 1:
import seaborn
import matplotlib.pyplot as plt
  
df = seaborn.load_dataset('tips')

graph = seaborn.FacetGrid(df, col ="sex",  hue ="day")

graph.map(plt.scatter, "total_bill", "tip", edgecolor ="w").add_legend()

plt.show()
```

```python
#Ejemplo 2:
import seaborn
import matplotlib.pyplot as plt
  
df = seaborn.load_dataset('tips')

graph = seaborn.FacetGrid(df, row ='smoker', col ='time')

graph.map(plt.hist, 'total_bill', bins = 15, color ='orange')

plt.show()
```

```python
#Ejemplo 3:

import seaborn
import matplotlib.pyplot as plt
  
df = seaborn.load_dataset('tips')
  
graph = seaborn.FacetGrid(df, col ='time', hue ='smoker')

graph.map(seaborn.regplot, "total_bill", "tip").add_legend()

plt.show()
```

El gráfico **PairGrid** en Seaborn es una herramienta versátil que permite visualizar las relaciones entre pares de variables en un conjunto de datos. 

Esta función crea una matriz de gráficos de dispersión y distribuciones para todas las combinaciones posibles de pares de variables numéricas en tus datos. 

Es útil para comprender las relaciones entre las variables y para identificar patrones, tendencias y posibles correlaciones en un conjunto de datos multivariado.

```python
#Veamos un ejemplo! 
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

# Crear un PairGrid
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)  # Histograma en la diagonal principal
g.map_offdiag(sns.scatterplot)  # Gráficos de dispersión fuera de la diagonal
plt.show()
```
