La función **heatmap** permite crear mapas de calor o gráficos de correlación en Python con seaborn. Puedes pasar un conjunto de datos 2D o un data frame de pandas. 

En caso de utilizar un data frame de pandas los nombres de las filas y las columnas serán usados como etiquetas para los ejes.

```python
import numpy as np
import seaborn as sns

# Simulación de datos
np.random.seed(1)
data = np.random.rand(10, 10)

sns.heatmap(data)
```

```python
#Celdas cuadradas
import numpy as np
import seaborn as sns

np.random.seed(1)
data = np.random.rand(10, 10)

sns.heatmap(data, square = True)
```

```python
#Borde
import numpy as np
import seaborn as sns

np.random.seed(1)
data = np.random.rand(10, 10)

sns.heatmap(data,
            linewidths = 0.75, 
            linecolor = "white") # Color por defecto
```

```python
#Anotaciones de texto
import numpy as np
import seaborn as sns

np.random.seed(1)
data = np.random.rand(10, 10)

sns.heatmap(data,
            annot = True,
            annot_kws = {'size': 6}) # Argumentos adicionales 
```

```python
#Personalizacion de color
import numpy as np
import seaborn as sns

np.random.seed(1)
data = np.random.rand(10, 10)

sns.heatmap(data,
            cmap = "Blues")
```
