```python
#Boxplot
import numpy as np
import seaborn as sns
import random

# Simulaci√≥n de datos
rng = np.random.RandomState(5)
variable = rng.normal(0, 1, size = 100)
random.seed(5)

grupo = random.choices(["G1", "G2", "G3"], k = 100)

grupo2 = random.choices(["A", "B"], k = 100)

df = {'variable': variable, 'grupo': grupo, 'grupo2': grupo2}
```

```python
#Boxplot - Horizontal
import seaborn as sns

# Box plot
sns.boxplot(x = variable)

# Equivalente a:
sns.boxplot(x = "variable", data = df)
```

```python
#Boxplot - Vertical
import seaborn as sns

# Box plot
sns.boxplot(y = variable)
```

```python
#Color de fondo de la caja
import seaborn as sns

# Box plot
sns.boxplot(x = variable,
            color = "orange")
```

```python
#Personalizacion avanzada
import seaborn as sns

# Box plot
sns.boxplot(x = variable,
            fliersize = 10, # Tama√±o de los at√≠picos
            linewidth = 3,  # Grosor de las l√≠neas
            saturation = 0.95) # Saturaci√≥n del color
```

```python
#Box plot por grupo
import seaborn as sns

# Box plot por grupo
sns.boxplot(x = variable,
            y = grupo)
```

```python
#Cambio de horientacion
import seaborn as sns

# Box plot por grupo
sns.boxplot(y = variable,
            x = grupo)
```

```python
#Ordenamiento
import seaborn as sns

# Box plot por grupo
sns.boxplot(y = variable,
            x = grupo,
            order = ["G1", "G2", "G3"])
```

```python
#Paleta de colores
import seaborn as sns

# Box plot por grupo
sns.boxplot(y = variable,
            x = grupo,
            palette = "Set3")
```

```python
#Box plot por grupo y subgrupo
import seaborn as sns

# Box plot por grupo
sns.boxplot(y = variable,
            x = grupo,
            hue = grupo2)
```

```python
#Violin
import seaborn as sns

sns.violinplot(x = variable)

# Equivalent to:
sns.violinplot(x = "variable", data = df)
```

```python
#Violin vertical
import seaborn as sns

sns.violinplot(y = variable)
```

```python
#Personalizacion del color
import seaborn as sns

sns.violinplot(x = variable,
               color = "green")
```

```python
#Recortar las colas de la distribuci√≥n
import seaborn as sns

sns.violinplot(x = variable,
               cut = 0)
```

```python
#Grafico de violin por grupo
import seaborn as sns

sns.violinplot(x = grupo, y = variable)
```

```python
#Color por grupo
import seaborn as sns

sns.violinplot(x = grupo, y = variable,
               hue = grupo2)
```

```python
#Por ultimo personalizacion de la paleta de colores
import seaborn as sns

sns.violinplot(x = grupo, y = variable, 
               palette = "Set2")
```
