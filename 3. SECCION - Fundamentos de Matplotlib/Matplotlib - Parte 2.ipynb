# Gráfico de Líneas

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, linestyle = 'solid') #default

plt.show()
```

```python
plt.plot(ypoints, linestyle = 'dashed')
```

```python
plt.plot(ypoints, linestyle = 'dotted')
```

```python
plt.plot(ypoints, linestyle = 'dashdot')
```

```python
plt.plot(ypoints, linestyle = 'None')
```

```python
#Hablemos ahora del color!
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = 'r')
plt.show()
```

```python
plt.plot(ypoints, c = '#4CAF50')
```

```python
plt.plot(ypoints, c = 'hotpink')
```

```python
#Ancho de la linea!
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20.5', c = 'hotpink')
plt.show()
```

```python
#Multiple lineas
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

# Personalizacion de Labels

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("eje x")
plt.ylabel("eje y")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Esto es un título de ejemplo")
plt.xlabel("eje x")
plt.ylabel("eje y")

plt.show()
```

```python
#Customización avanzada!
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Esto es un titulo de ejemplo", fontdict = font1) #loc = 'left'
plt.xlabel("eje X", fontdict = font2)
plt.ylabel("eje Y", fontdict = font2)

plt.plot(x, y)
plt.show()
```
