```python
#ejemplo 1
# Datos para el gráfico
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Crear un gráfico de línea
plt.plot(x, y)

# Mostrar el gráfico
plt.show()
```

```python
#ejemplo 2:
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 10])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
#ejemplo 3:
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
#Que pasa si por ej, quisieramos colocar un punto en el (1,8) y (3,15) ?
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 15])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
#Veamos ahora múltiples lineas!
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
#Para finalizar markers
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 6, 1, 12])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
plt.plot(ypoints, marker = '*')
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 6, 1, 15])

plt.plot(ypoints, marker = 'o', ms = 30)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 6, 1, 15])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'hotpink', mfc = 'hotpink')
```

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
```
