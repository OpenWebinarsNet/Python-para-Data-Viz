# Scatter plot
-------------
1. Crear un scatterplot entre x='total_bill' - y='tip' a través del uso de la Librería Seaborn:


```python
sns.scatterplot(x='total_bill', y='tip', data = df)
plt.show()
```

2. Modificar el tamaño del plot por ancho = 20, alto= 10. Tip: Hay que utilizar plt.figure:

```python
plt.figure(figsize= (20,10))
sns.scatterplot(x='total_bill', y='tip', data = df)
```

3. Cambiar el color por defecto que tiene el grafico por el color verde:

```python
plt.figure(figsize= (20,10))
sns.scatterplot(x='total_bill', y='tip', data = df, color = 'green')
plt.show()
```

4. Asignar un color en base a la variable "sex". Tip: Hay que utilizar el atributo - hue:

```python
plt.figure(figsize= (20,10))
sns.scatterplot(x='total_bill', y='tip', data = df, hue = 'sex')
plt.show()
```

5. Asignar un forma diferente a cada point en base a los valores de la variable denominada "smoker". Tip: Hay que utilizar el atributo - style:

```python
plt.figure(figsize= (20,10))
sns.scatterplot(x='total_bill', y='tip', data = df, hue = 'sex', style = 'smoker')
plt.show()
```

**Extra**: Agregar a nuestro gráfico los siguientes labels:

* Titulo: "Importe del Tickets vs Propinas"
* Eje x: "Importe del Tickets"
* Eje y: "Propinas"


```python
p={'Female':'blue','Male':'green'}

plt.figure(figsize= (20,10))
sns.scatterplot(x='total_bill', y='tip', data = df, hue = 'sex', style = 'smoker', palette = p)
plt.xlabel('Importe del Tickets')
plt.ylabel('Propinas')
plt.title('Importe del Tickets vs Propinas')
plt.show()
```

# Distplot

1. Generar un histograma para la variable total_bill:


```python
plt.figure(figsize=(20,10))
sns.distplot(df.total_bill, kde=True, hist=True, bins = 15)
plt.xlabel('Total Bill')
plt.ylabel('Frecuencia')
plt.title('Distribucion de Total Bill')
```

2. Comparar en una misma gráfica las distribuciones de total_bill y tip: 

```python
fig = plt.figure(figsize=(20,10))

sns.distplot(df.total_bill, kde=True, hist=True, bins = 15, label='Total Bill')
sns.distplot(df.tip, kde=True, hist=True, bins = 15, label = 'Tips')

fig.legend()
plt.xlabel('Valor')
plt.ylabel('Frecuencias')
plt.title('Distribucions')
```

# Boxplot
